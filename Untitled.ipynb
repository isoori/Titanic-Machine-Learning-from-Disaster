```python
dff= pd.read_csv("test.csv")
dff.head()
```

```python
df = df.drop(['Ticket', 'Cabin'], axis=1)
dff = dff.drop(['Ticket', 'Cabin'], axis=1)
combine = [df, dff]
```

```python
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Z a-z]+)\.', expand=False)
```

```python
df
```

```python
dff
```

```python
for dataset in combine:
    dataset['Title'] = dataset['Title'].replace(['Dr','Rev','Major','Col','the Countess','Capt','Lady','Don','Jonkheer'],'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle','Miss')
    dataset['Title'] = dataset['Title'].replace('Ms','Miss')
    dataset['Title'] = dataset['Title'].replace('Mme','Mrs')
    dataset['Title'] = dataset['Title'].replace("Sir",'Mr')
```

```python
df['Title'].value_counts()
```

```python
mapping = {'Mr':1,'Miss':2,'Mrs':3,'Master':4,'Rare':5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(mapping)
    
```

```python
df['Title'].fillna(0,inplace=True)
```

```python
dff['Title'].fillna(0,inplace=True)
```

```python
dff.set_index('PassengerId',inplace=True)
```

```python
df
```

```python
dff
```

```python
for dataset in combine:
    dataset.drop("Name",axis=1,inplace=True)
```

```python
for dataset in combine:    
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age'] = 4
```

```python
for dataset in combine:
    dataset['Age'].fillna(dataset['Age'].median(),inplace=True)
```

```python
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map({'male':0, 'female':1}).astype(int)
```

```python
df
```

```python
dff
```

```python
for dataset in combine:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
```

```python
for dataset in combine:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize']==1,'IsAlone'] = 1
```

```python
for dataset in combine:    
    dataset.loc[ dataset['Fare'] <= 8, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 8) & (dataset['Fare'] <= 16), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 16) & (dataset['Fare'] <= 32), 'Fare'] = 2
    dataset.loc[(dataset['Fare'] > 32),'Fare'] = 3
```

```python
df
```

```python
for dataset in combine:
    dataset.drop(['SibSp','Parch'],axis=1,inplace=True)
```

```python
df
```

```python
dff
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map({'S':0,'C':1,'Q':2})
```

```python
df
```

```python
dff
```

```python
df['Embarked'].fillna(0,inplace=True)
```

```python
dff['Embarked'].fillna(0,inplace=True)
```

```python
dff['Fare'].fillna(0,inplace=True)
```

```python
features = df.drop('Survived',axis=1)
target = df['Survived']
```

```python
lg = LogisticRegression()
lg.fit(features,target)
prediction = lg.predict(dff)
F1 = f1_score(TEST['Survived'],prediction)
PRECISION = f1_score(TEST['Survived'],prediction)
RECALL = recall_score(TEST['Survived'],prediction)
ACCURACY = accuracy_score(TEST['Survived'],prediction)
m1= pd.DataFrame({'F1':F1,'PRECISION':PRECISION,'ACCURACY':ACCURACY,"RECALL":RECALL},index=['LOGISTIC_REGRESSION'])
m1
```

```python
tree = DecisionTreeClassifier()
tree.fit(features,target)
prediction = tree.predict(dff)
F1 = f1_score(TEST['Survived'],prediction)
PRECISION = f1_score(TEST['Survived'],prediction)
RECALL = recall_score(TEST['Survived'],prediction)
ACCURACY = accuracy_score(TEST['Survived'],prediction)
m2= pd.DataFrame({'F1':F1,'PRECISION':PRECISION,'ACCURACY':ACCURACY,"RECALL":RECALL},index=['DECISION_TREE'])
m2
```

```python
forest = RandomForestClassifier()
forest.fit(features,target)
prediction = forest.predict(dff)
F1 = f1_score(TEST['Survived'],prediction)
PRECISION = f1_score(TEST['Survived'],prediction)
RECALL = recall_score(TEST['Survived'],prediction)
ACCURACY = accuracy_score(TEST['Survived'],prediction)
m3= pd.DataFrame({'F1':F1,'PRECISION':PRECISION,'ACCURACY':ACCURACY,"RECALL":RECALL},index=['RandomForestClassifier'])
m3
```

```python
xgb = XGBClassifier()
xgb.fit(features,target)
prediction = xgb.predict(dff)
F1 = f1_score(TEST['Survived'],prediction)
PRECISION = f1_score(TEST['Survived'],prediction)
RECALL = recall_score(TEST['Survived'],prediction)
ACCURACY = accuracy_score(TEST['Survived'],prediction)
m4= pd.DataFrame({'F1':F1,'PRECISION':PRECISION,'ACCURACY':ACCURACY,"RECALL":RECALL},index=['XGBClassifier'])
m4
```

```python
knn = KNeighborsClassifier()
knn.fit(features,target)
prediction = knn.predict(dff)
F1 = f1_score(TEST['Survived'],prediction)
PRECISION = f1_score(TEST['Survived'],prediction)
RECALL = recall_score(TEST['Survived'],prediction)
ACCURACY = accuracy_score(TEST['Survived'],prediction)
m5= pd.DataFrame({'F1':F1,'PRECISION':PRECISION,'ACCURACY':ACCURACY,"RECALL":RECALL},index=['KNeighborsClassifier'])
m5
```

```python
result = pd.concat([m1,m2,m3,m4,m5],axis=0)
result
```
